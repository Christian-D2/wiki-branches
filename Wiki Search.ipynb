```python
import requests
from bs4 import BeautifulSoup

def get_wikipedia_links(url):
    # Send a GET request to the Wikipedia page
    response = requests.get(url)
    
    # Create a BeautifulSoup object from the HTML content
    soup = BeautifulSoup(response.content, 'html.parser')
    
    # Find all the anchor tags with 'href' attribute
    link_tags = soup.find_all('a', href=True)
    
    # Extract the href attribute from each anchor tag
    links = [link['href'] for link in link_tags]
    
    # Filter out non-Wikipedia links
    wikipedia_links = [link for link in links if link.startswith('/wiki/')]
    
    return wikipedia_links

# Example usage
url = 'https://en.wikipedia.org/wiki/Drug_policy_of_South_Korea'
links = get_wikipedia_links(url)

#print(links[24])

# Print the links
for link in links:
    link = 'https://en.wikipedia.org' + link
    #print(link)
    if(link == 'https://en.wikipedia.org/wiki/Gangnam_District'):
        print('we found it')
        
```

```python
import requests

def get_wikipedia_summary(url):
    # Extract the page title from the URL
    page_title = url.split("/")[-1]

    # API endpoint for Wikipedia summary
    api_endpoint = f"https://en.wikipedia.org/api/rest_v1/page/summary/{page_title}"

    try:
        # Send GET request to the API
        response = requests.get(api_endpoint)
        data = response.json()

        # Extract the summary from the response
        summary = data["extract"]

        return truncate_string(summary, 60)

    except requests.exceptions.RequestException as e:
        print(f"An error occurred: {e}")

def truncate_string(input_string, max_words):
    # Split the string into words
    words = input_string.split()

    # Check if the number of words is less than or equal to max_words
    if len(words) <= max_words:
        return ' '.join(words)  # Return the original string

    # Otherwise, take the first max_words words and add '...' to the end
    truncated_words = words[:max_words]
    truncated_string = ' '.join(truncated_words) + '...'

    return truncated_string        
        
# Example usage
#url = "https://en.wikipedia.org/wiki/Python_(programming_language)"
#summary = get_wikipedia_summary(url)
#print(summary)
```
